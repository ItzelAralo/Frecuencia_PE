```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
Lb = 0.85 #lambda de nacimiento 
Ld = 0.10 #lambda de muerte 
Nt = 30 #pasos  numero de eventos que voy a medir, cada uno va a estar separado por un tiempo
x = 10 # aqui voy a poner mi poblacion a Enero
```

```python
#vamos a probar si el proceso sube 1 o baja 1 
```

```python
T = [] #aqui voy a ir almacenando los tiempos, tiempo  que hay entre tiempos temporales 
X_t = [x] #aqui voy a ir almacenando mi poblacion
for i in range(Nt):
    Tb = np.random.exponential(Lb)
    Td = np.random.exponential(Ld)
    if (Tb>Td) | (x == 0):
        x = x + 1 
        T.append(Tb)
    else : 
        x = x - 1
        T.append(Td)
    X_t.append(x)
```
